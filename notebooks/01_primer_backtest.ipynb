## Primer Backtest: Descargar y Visualizar Datos

**Documento de referencia:** `04_ruta_implementacion_fase0_fase1.md` — Sección 4.0.4

**Objetivo:** Romper la inercia inicial. Obtener datos financieros reales y visualizarlos.
Si ves un gráfico del SPY, los datos fluyen correctamente.

````python
# 4.0.4 — Primer script: descargar y visualizar datos
import yfinance as yf
import matplotlib.pyplot as plt

# Descargar 5 años de datos diarios del SPY
df = yf.download("SPY", start="2019-01-01", end="2024-01-01")

# Mostrar las primeras filas para ver la estructura
print(df.head())
print(f"\nFilas: {len(df)} | Columnas: {list(df.columns)}")
print(f"Desde: {df.index[0].date()}  Hasta: {df.index[-1].date()}")
````

````python
# Graficar el precio de cierre ajustado
df['Adj Close'].plot(figsize=(12, 6), title='SPY — Precio de Cierre Ajustado', linewidth=1.5)
plt.ylabel('Precio (USD)')
plt.xlabel('Fecha')
plt.grid(True, alpha=0.3)
plt.show()

print("✅ Gráfico generado. Los datos fluyen correctamente.")
````
